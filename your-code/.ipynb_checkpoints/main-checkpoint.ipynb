## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

```python
# your code here
orders.columns
```

```python
orders['CustomerID'].nunique()
```

```python
orders['amount_spent'] = orders['amount_spent'].astype(int)
orders['amount_spent'].dtype
```

```python
customer_spend = (orders.groupby('CustomerID', as_index=False)['amount_spent'].sum())
customer_spend
```

```python
low_q = customer_spend['amount_spent'].quantile(0.75)
high_q = customer_spend['amount_spent'].quantile(0.95)
```

```python
vip_customers = customer_spend[(customer_spend['amount_spent'] > high_q)]
preffered_customers = customer_spend[(customer_spend['amount_spent'] >= low_q) & (customer_spend['amount_spent'] <= high_q)]
```

```python
vip_customers.shape
```

```python
customer_spend.shape
```

```python
conditions = [(customer_spend['amount_spent'] > high_q), 
              (customer_spend['amount_spent'] >= low_q) & (customer_spend['amount_spent'] <= high_q)]

labels = ["VIP", "Preffered"]

customer_spend["VIP/Preffered"] = np.select(conditions, labels, default="Regular")
```

```python
customer_spend
```

Now we'll leave it to you to solve Q2 & Q3, which you can leverage from your solution for Q1:

## Q2: How to identify which country has the most VIP Customers?

```python
# your code here
merged_cust = pd.merge(left = customer_spend, 
                       right = orders,
                       how = 'inner',
                       left_on = "CustomerID",
                       right_on = "CustomerID"
)

merged_cust
```

```python
vip_country_counts = (
    merged_cust[merged_cust['VIP/Preffered'] == 'VIP']
    .groupby('Country')['CustomerID']
    .nunique()
    .sort_values(ascending=False)
)
vip_country_counts.head()
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# your code here
combined_counts = (
    merged_cust[(merged_cust['VIP/Preffered'] == 'VIP') | (merged_cust['VIP/Preffered'] == 'Preffered')]
    .groupby('Country')['CustomerID']
    .nunique()
    .sort_values(ascending=False)
)
combined_counts.head()
```
